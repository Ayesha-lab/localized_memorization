```python
# Specify the file path in your Google Drive
file_path = '/content/drive/MyDrive/noisy_gradients.csv'

# Read the CSV file into a DataFrame
noisy_gradients = pd.read_csv(file_path)

# Display the DataFrame
noisy_gradients
```

```python
merged_df = pd.merge(noisy_gradients[['name', 'average']], clean_gradients[['name', 'average']], on='name', suffixes=('_noisy', '_clean'))

# Rename columns for clarity
merged_df.rename(columns={'average_noisy': 'noisy_avg', 'average_clean': 'clean_avg'}, inplace=True)

# Display the resulting DataFrame
print(merged_df)
```

```python
import matplotlib.pyplot as plt

# Assuming merged_df is the merged DataFrame from the previous example
x_values = range(1, len(merged_df) + 1)  # Using a default numeric index
y_noisy = merged_df['noisy_avg']
y_clean = merged_df['clean_avg']

# Plotting
plt.plot(x_values, y_noisy, label='Noisy Average')
plt.plot(x_values, y_clean, label='Clean Average')

# Adding labels and title
plt.xlabel('Index')
plt.ylabel('Average')
plt.title('Noisy vs Clean Averages')

# Adding a legend
plt.legend()

# Show the plot
plt.show()
```
